```python
import duckdb
import pandas as pd
from pathlib import Path
import xgboost as xgb
from sklearn.metrics import mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data_dir = Path("data")
db_path = data_dir / "news_therms.duckdb"
con = duckdb.connect(db_path)

query = """
WITH dense_terms AS (
    SELECT term
    FROM term_occurrence
    WHERE term_occurrence.source IN ('UP', 'RBC')
    GROUP BY term
    HAVING count(*) > 100
    ORDER BY count(*) DESC
    LIMIT 2000
),
daily_counts AS (
    SELECT
        term,
        CAST(date AS DATE) AS usage_date,
        count(*) AS daily_count
    FROM term_occurrence
    WHERE term IN (SELECT term FROM dense_terms)
      AND source IN ('UP', 'RBC')
    GROUP BY term, usage_date
)
SELECT * FROM daily_counts
ORDER BY term, usage_date ASC;
"""

df_raw = con.execute(query).df()
```

```python
df_raw['usage_date'] = pd.to_datetime(df_raw['usage_date'])

df_filled = (
    df_raw.pivot(index='usage_date', columns='term', values='daily_count')
    .fillna(0)
    .asfreq('D', fill_value=0) # Ensures every single day in the range exists
    .reset_index()
    .melt(id_vars='usage_date', value_name='daily_count')
    .sort_values(by=['term', 'usage_date'])
    .reset_index(drop=True)
)
```

```python
lags = [1, 2, 3, 7, 14]
for lag in lags:
    df_filled[f'lag_{lag}'] = df_filled.groupby('term')['daily_count'].shift(lag)

df_filled['rolling_3d_mean'] = df_filled.groupby('term')['lag_1'].transform(lambda x: x.rolling(window=3).mean())
df_filled['rolling_7d_mean'] = df_filled.groupby('term')['lag_1'].transform(lambda x: x.rolling(window=7).mean())

df_filled['diff_1d'] = df_filled.groupby('term')['lag_1'].transform(lambda x: x.diff())

df_filled = df_filled.dropna().reset_index(drop=True)
```

```python
df_filled
```

```python
max_date = df_filled['usage_date'].max()
test_start_date = max_date - pd.Timedelta(days=13)
val_start_date = test_start_date - pd.Timedelta(days=14)
```

```python
df_train = df_filled[df_filled['usage_date'] < val_start_date].copy()
df_val = df_filled[(df_filled['usage_date'] >= val_start_date) & (df_filled['usage_date'] < test_start_date)].copy()
df_test = df_filled[df_filled['usage_date'] >= test_start_date].copy()

print(f"Total date range: {df_filled['usage_date'].min().date()} to {max_date.date()}")
print(f"Training shape:   {df_train.shape} (Ends {val_start_date.date() - pd.Timedelta(days=1)})")
print(f"Validation shape: {df_val.shape} (Ends {test_start_date.date() - pd.Timedelta(days=1)})")
print(f"Testing shape:    {df_test.shape} (Starts {test_start_date.date()})")
```

```python
target = 'daily_count'
features = [col for col in df_filled.columns if col not in ['term', 'usage_date', target]]
```

```python
X_train, y_train = df_train[features], df_train[target]
X_val,   y_val   = df_val[features],   df_val[target]
X_test,  y_test  = df_test[features],  df_test[target]

model = xgb.XGBRegressor(
    n_estimators=1000,
    learning_rate=0.05,
    max_depth=5,
    objective='count:poisson',
    early_stopping_rounds=50,
    random_state=42
)

model.fit(
    X_train, y_train,
    eval_set=[(X_train, y_train), (X_val, y_val)],
    verbose=100
)
```

```python
predictions = model.predict(X_test)

df_results = df_test.copy()
df_results['predicted_count'] = predictions

mae = mean_absolute_error(df_results['daily_count'], df_results['predicted_count'])
mse = mean_squared_error(df_results['daily_count'], df_results['predicted_count'])

print(f"MAE = {mae:.2f}")
print(f"MSE = {mse:.2f}")
```

```python
top_3_terms = (
    df_results.groupby('term')['daily_count']
    .sum()
    .nlargest(3)
    .index.tolist()
)
print(f"Analyzing Top 3 Terms: {top_3_terms}")

df_top3 = df_results[df_results['term'].isin(top_3_terms)]

plt.figure(figsize=(14, 7))
sns.set_theme(style="whitegrid")

colors = sns.color_palette("Set1", n_colors=3)
term_color_map = dict(zip(top_3_terms, colors))

for term in top_3_terms:
    term_data = df_top3[df_top3['term'] == term]

    # Plot Actual Usage (Solid line with dots)
    sns.lineplot(
        x=term_data['usage_date'],
        y=term_data['daily_count'],
        label=f"{term} (Actual)",
        color=term_color_map[term],
        marker='o',
        linewidth=2
    )

    # Plot Predicted Usage (Dashed line with 'x' markers)
    sns.lineplot(
        x=term_data['usage_date'],
        y=term_data['predicted_count'],
        label=f"{term} (Predicted)",
        color=term_color_map[term],
        linestyle='--',
        marker='X',
        alpha=0.7
    )

plt.title('Прогнозована та реальна кількість термінів за останні два тижні', fontsize=16)
plt.xlabel('Дата', fontsize=12)
plt.ylabel('Кількість', fontsize=12)
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
history_df = df_filled[df_filled['term'].isin(top_3_terms)].copy()
history_df = history_df.sort_values(by=['term', 'usage_date']).reset_index(drop=True)

future_days = 7
last_known_date = history_df['usage_date'].max()

for i in range(1, future_days + 1):
    next_date = last_known_date + pd.Timedelta(days=i)

    new_rows = pd.DataFrame({
        'term': top_3_terms,
        'usage_date': [next_date] * len(top_3_terms),
        'daily_count': [np.nan] * len(top_3_terms) # To be filled by XGBoost
    })

    history_df = pd.concat([history_df, new_rows], ignore_index=True)
    history_df = history_df.sort_values(by=['term', 'usage_date']).reset_index(drop=True)

    for lag in [1, 2, 3, 7, 14]:
        history_df[f'lag_{lag}'] = history_df.groupby('term')['daily_count'].shift(lag)

    history_df['rolling_3d_mean'] = history_df.groupby('term')['lag_1'].transform(lambda x: x.rolling(window=3).mean())
    history_df['rolling_7d_mean'] = history_df.groupby('term')['lag_1'].transform(lambda x: x.rolling(window=7).mean())
    history_df['diff_1d']         = history_df.groupby('term')['lag_1'].transform(lambda x: x.diff())

    predict_set = history_df[history_df['usage_date'] == next_date].copy()
    X_predict = predict_set[features]
    preds = model.predict(X_predict)

    for term, pred in zip(top_3_terms, preds):
        mask = (history_df['term'] == term) & (history_df['usage_date'] == next_date)
        history_df.loc[mask, 'daily_count'] = max(0, round(pred))

forecast_df = history_df[history_df['usage_date'] > last_known_date].copy()
```

```python
history_window_days = 7
history_start = last_known_date - pd.Timedelta(days=history_window_days - 1)

history_plot_df = df_filled[
    (df_filled['term'].isin(top_3_terms)) &
    (df_filled['usage_date'] >= history_start) &
    (df_filled['usage_date'] <= last_known_date)
].copy()

forecast_plot_df = forecast_df.copy()

plt.figure(figsize=(14, 7))
sns.set_theme(style="whitegrid")

colors = sns.color_palette("Set1", n_colors=len(top_3_terms))
term_color_map = dict(zip(top_3_terms, colors))

for term in top_3_terms:
    term_hist = history_plot_df[history_plot_df['term'] == term].sort_values('usage_date')
    term_fore = forecast_plot_df[forecast_plot_df['term'] == term].sort_values('usage_date')

    sns.lineplot(
        x=term_hist['usage_date'],
        y=term_hist['daily_count'],
        label=f"{term} (History)",
        color=term_color_map[term],
        marker='o',
        linewidth=2
    )

    sns.lineplot(
        x=term_fore['usage_date'],
        y=term_fore['daily_count'],
        label=f"{term} (Forecast)",
        color=term_color_map[term],
        linestyle='--',
        marker='X',
        alpha=0.7
    )

plt.title('Прогноз на майбутнє (Останні 7 днів та наступні 7)', fontsize=16)
plt.xlabel('Дата', fontsize=12)
plt.ylabel('Кількість', fontsize=12)
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
