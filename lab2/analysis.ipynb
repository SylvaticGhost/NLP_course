```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
POS_TO_SKIP = ['PART', 'PUNCT', 'CCONJ', 'SCONJ', 'SYM', 'AUX', 'ADP', 'INTJ', 'DET', 'PRON']
df = pd.read_csv("words-not-translated.csv")
df = df[~df['upos'].isin(POS_TO_SKIP)]
df
```

```python
word_distribution = df.groupby('word').size().reset_index(name='frequency')
word_distribution = word_distribution.sort_values('frequency', ascending=False)
word_distribution
```

```python
words_same_across_languages = df[df.groupby('word')['language'].transform('nunique') > 1].drop('source', axis=1).groupby(['word', 'upos']).size().reset_index(name='count').sort_values('count', ascending=False)
words_same_across_languages
```

```python
pivot_df = words_same_across_languages.pivot_table(
    index='word',
    columns='upos',
    values='count',
    fill_value=0
)

for col in ['PROPN', 'NOUN', 'X']:
    if col not in pivot_df.columns:
        pivot_df[col] = 0

valid_words = pivot_df[
    (pivot_df['PROPN'] > pivot_df['NOUN']) &
    (pivot_df['PROPN'] > pivot_df['X'])
].index

words_same_across_languages = words_same_across_languages[words_same_across_languages['upos'].isin(['NOUN', 'X', 'PROPN'])]
final_df = words_same_across_languages[words_same_across_languages['word'].isin(valid_words)]
final_df = final_df.groupby('word').agg({'count': 'sum'}).reset_index().sort_values(by='count', ascending=False)

manual_filters = ['Science', 'Vision', 'Team', 'Business']
final_df = final_df[~final_df['word'].isin(manual_filters)]

final_df
```

```python
plt.figure(figsize=(12, 8))
plt.barh(final_df['word'][:20], final_df['count'][:20], color='skyblue')
plt.title('Top 20 Technologies by Frequency', fontsize=16)
plt.xlabel('Usage Count', fontsize=12)
plt.ylabel('Technology', fontsize=12)
plt.tight_layout()
plt.show()
```
