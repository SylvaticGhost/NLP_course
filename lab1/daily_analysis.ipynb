```python
import duckdb
import pandas as pd
from pathlib import Path
```

```python
START_DATE = '2026-02-07'
END_DATE = '2026-02-13'
```

```python
data_dir = Path("data")
db_path = data_dir / "news_therms.duckdb"


query = f"""
WITH day_part_terms AS (
    SELECT
        DATE_TRUNC('day', date) AS day,
        CASE
            WHEN EXTRACT(HOUR FROM date) >= 6 AND EXTRACT(HOUR FROM date) < 12 THEN 'morning'
            WHEN EXTRACT(HOUR FROM date) >= 12 AND EXTRACT(HOUR FROM date) < 18 THEN 'afternoon'
            WHEN EXTRACT(HOUR FROM date) >= 18 AND EXTRACT(HOUR FROM date) < 24 THEN 'evening'
            ELSE 'night'
        END AS day_part,
        term,
        source,
        COUNT(*) AS occurrence_count
    FROM term_occurrence
    WHERE date >= '{START_DATE}' AND date < '{END_DATE}'
      AND source IN ('RBC', 'UP')
    GROUP BY DATE_TRUNC('day', date), day_part, term, source
),
ranked_terms AS (
    SELECT
        day,
        day_part,
        term,
        source,
        occurrence_count,
        ROW_NUMBER() OVER (PARTITION BY day, day_part, source ORDER BY occurrence_count DESC, term) AS rank
    FROM day_part_terms
)
SELECT
    day,
    day_part,
    source,
    term,
    occurrence_count,
    rank
FROM ranked_terms
WHERE rank <= 5
ORDER BY day,
         CASE day_part
             WHEN 'morning' THEN 1
             WHEN 'afternoon' THEN 2
             WHEN 'evening' THEN 3
             WHEN 'night' THEN 4
         END,
         source,
         rank
"""

with duckdb.connect(str(db_path)) as con:
    df = con.execute(query).df()

con.close()
```

## Vizualization

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import matplotlib

matplotlib.rcParams['font.family'] = 'DejaVu Sans'

TERM_COL = 'Топ 5' if 'Топ 5' in df.columns else 'term'
FREQ_COL = 'Частота' if 'Частота' in df.columns else 'occurrence_count'

word_freq = {}
for _, row in df.iterrows():
    term = row[TERM_COL]
    freq = row[FREQ_COL]
    if term in word_freq:
        word_freq[term] += freq
    else:
        word_freq[term] = freq

wordcloud = WordCloud(
    width=1200,
    height=600,
    background_color='white',
    colormap='viridis',
    relative_scaling=0.5,
    min_font_size=10,
    font_path='C:/Windows/Fonts/arial.ttf'
).generate_from_frequencies(word_freq)

plt.figure(figsize=(15, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Топ слів (2026-02-07 to 2026-02-13)', fontsize=16, fontweight='bold')
plt.tight_layout(pad=0)
plt.savefig('wordcloud_top5_terms.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
sources = df['source'].unique()

fig, axes = plt.subplots(1, len(sources), figsize=(18, 6))
if len(sources) == 1:
    axes = [axes]

for idx, source in enumerate(sources):
    source_df = df[df['source'] == source]

    source_word_freq = {}
    for _, row in source_df.iterrows():
        term = row[TERM_COL]
        freq = row[FREQ_COL]
        if term in source_word_freq:
            source_word_freq[term] += freq
        else:
            source_word_freq[term] = freq

    wc = WordCloud(
        width=600,
        height=400,
        background_color='white',
        colormap='plasma',
        relative_scaling=0.5,
        min_font_size=8,
        font_path='C:/Windows/Fonts/arial.ttf'
    ).generate_from_frequencies(source_word_freq)

    # Display
    axes[idx].imshow(wc, interpolation='bilinear')
    axes[idx].axis('off')
    axes[idx].set_title(f'{source} - Топ слів', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.savefig('wordcloud_by_source.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
term_totals = df.groupby(TERM_COL)[FREQ_COL].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
bars = plt.barh(range(len(term_totals)), term_totals.values, color='steelblue')
plt.yticks(range(len(term_totals)), term_totals.index)
plt.xlabel('Total Frequency', fontsize=12)
plt.ylabel('Терм', fontsize=12)
plt.title('Топ 10 найчастіших термів', fontsize=14, fontweight='bold')
plt.gca().invert_yaxis()

# Add value labels on bars
for i, (term, value) in enumerate(term_totals.items()):
    plt.text(value + 0.5, i, str(int(value)), va='center', fontsize=10)

plt.tight_layout()
plt.savefig('top10_terms_barchart.png', dpi=300, bbox_inches='tight')
plt.show()
```
