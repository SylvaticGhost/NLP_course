```python
import duckdb
import json
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
with open('./models/stanza_cache.json', 'r', encoding='utf-8') as f:
    cache = json.load(f)

words = cache['words']
filtered_words = cache['filtered_words']

con = duckdb.connect("./data/products.duckdb")
query = """
    SELECT title, text
    FROM reviews
    WHERE (text IS NOT NULL AND text != '')
       OR (title IS NOT NULL AND title != '')
"""
df = con.execute(query).df()
con.close()

df['title'] = df['title'].fillna('')
df['text'] = df['text'].fillna('')
df['combined_text'] = df['title'] + " " + df['text']
```

```python
print(f'Кількість дослідженних коментарів: {df.shape[0]}')
```

## TF-IDF аналіз

```python
vectorizer = TfidfVectorizer(stop_words='english', max_features=1000)
tfidf_matrix = vectorizer.fit_transform(df['combined_text'])

sum_tfidf = tfidf_matrix.sum(axis=0)
tfidf_scores = [(word, sum_tfidf[0, idx]) for word, idx in vectorizer.vocabulary_.items()]
tfidf_scores = sorted(tfidf_scores, key=lambda x: x[1], reverse=True)

print("Топ-15 слів за вагою TF-IDF:")
for word, score in tfidf_scores[:15]:
    print(f"{word}: {score:.2f}")
```

## Лексична дисперсія

```python
target_words = ['coffee', 'hot', 'cold', 'brew', 'water', 'machine']

dispersion_data = {word: [] for word in target_words}
for i, word in enumerate(words):
    if word in dispersion_data:
        dispersion_data[word].append(i)

plt.figure(figsize=(12, 6))
plt.title("Лексична дисперсія ключових слів")
for i, word in enumerate(target_words):
    plt.vlines(dispersion_data[word], i - 0.4, i + 0.4, colors='blue', alpha=0.5)

plt.yticks(range(len(target_words)), target_words)
plt.ylim(-1, len(target_words))
plt.xlabel("Позиція слова в корпусі тексту")
plt.show()
```

## Розподіл довжини слів

```python
word_lengths = [len(word) for word in filtered_words]

plt.figure(figsize=(10, 5))
sns.histplot(word_lengths, bins=range(1, 15), kde=True, color='skyblue')
plt.title('Розподіл довжини слів у відгуках')
plt.xlabel('Довжина слова (кількість літер)')
plt.ylabel('Частота')
plt.xticks(range(1, 15))
plt.grid(axis='y', alpha=0.75)
plt.show()
```

## Найчастіші біграми

```python
bigrams = list(zip(filtered_words[:-1], filtered_words[1:]))
bigram_freq = Counter(bigrams)

print("Топ-10 найчастіших біграм:")
for bigram, freq in bigram_freq.most_common(10):
    print(f"{bigram[0]} {bigram[1]}: {freq} разів")
```
