```python
import duckdb
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
```

```python
START_DATE = '2026-01-31'
END_DATE = '2026-02-13'
```

```python
data_dir = Path("data")
db_path = data_dir / "news_therms.duckdb"

query = f"""
WITH top_terms AS (
    SELECT term
    FROM term_occurrence
        WHERE CAST(date AS DATE) >= '{START_DATE}' AND CAST(date AS DATE) <= '{END_DATE}'
        AND source in ('UP', 'RBC')
    GROUP BY term
    ORDER BY count(*) DESC
    LIMIT 3
)
SELECT
    term,
    CAST(date AS DATE) AS usage_date,
    count(*) AS daily_count
FROM term_occurrence
    WHERE CAST(date AS DATE) >= '{START_DATE}' AND CAST(date AS DATE) <= '{END_DATE}'
  AND term IN (SELECT term FROM top_terms)
GROUP BY term, usage_date
ORDER BY usage_date ASC;
"""

with duckdb.connect(str(db_path)) as con:
    df = con.execute(query).df()

con.close()
```

```python
df
```

```python
plt.figure(figsize=(12, 6))
sns.set_theme(style="whitegrid")

sns.barplot(
    data=df,
    x='usage_date',
    y='daily_count',
    hue='term',
    palette='Set2'
)

plt.title(f'топ 3 терміни з {START_DATE} до {END_DATE}', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Daily Occurrences', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.set_theme(style="whitegrid")

sns.lineplot(
    data=df,
    x='usage_date',
    y='daily_count',
    hue='term',
    marker='o',
    linewidth=2
)

plt.title(f'Daily Usage of Top 3 Terms ({START_DATE} to {END_DATE})', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Daily Occurrences', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
