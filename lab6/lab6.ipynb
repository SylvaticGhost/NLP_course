```python
import pandas as pd
from transformers import pipeline
from tqdm import tqdm
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
from sentence_transformers import SentenceTransformer
```

```python
tqdm.pandas(desc="Нейромережевий аналіз тональності")
df = pd.read_csv("./../lab5/ecommerce_unified_dataset_clean.csv")
```

```python
from lab6.loadModels import DISTILBERT_PATH

print("Завантаження моделі DistilBERT...")
sentiment_pipeline = pipeline(
    "sentiment-analysis",
    model=DISTILBERT_PATH,
    tokenizer=DISTILBERT_PATH,
    truncation=True,
    max_length=512
)
```

```python
def get_nn_sentiment(text):
    try:
        result = sentiment_pipeline(str(text))[0]
        score = result['score']
        return score if result['label'] == 'POSITIVE' else -score
    except Exception:
        return 0.0

print(f"Починаємо аналіз {len(df)} текстів")
df['nn_sentiment_score'] = df['text'].progress_apply(get_nn_sentiment)
```

```python
summary_nn = df.groupby('platform')['nn_sentiment_score'].mean().round(3)
summary_nn
```

```python
from lab6.loadModels import MINILM_LOCAL_PATH

embedding_model = SentenceTransformer(MINILM_LOCAL_PATH)
embeddings = embedding_model.encode(df['text'].astype(str).tolist(), show_progress_bar=True)
```

```python
pca = PCA(n_components=2)
pca_result = pca.fit_transform(embeddings)

df['pca_x'] = pca_result[:, 0]
df['pca_y'] = pca_result[:, 1]
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(
    data=df,
    x='pca_x',
    y='pca_y',
    hue='platform',
    palette='Set1',
    alpha=0.7,
    s=60
)
plt.title('Нейромережевий семантичний аналіз описів товарів (Sentence Transformers + PCA)', fontsize=14)
plt.xlabel('Головна компонента 1 (Семантична вісь X)')
plt.ylabel('Головна компонента 2 (Семантична вісь Y)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```
