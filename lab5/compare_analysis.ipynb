```python
import pandas as pd
from textstat.textstat import textstat
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
df = pd.read_csv("ecommerce_unified_dataset_clean.csv")
```

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
def get_vader_score(text):
    return analyzer.polarity_scores(str(text))['compound']

def get_lexical_diversity(text):
    words = str(text).split()
    if not words: return 0
    return len(set(words)) / len(words)
```

```python
df['sentiment_score'] = df['text'].apply(get_vader_score)
df['readability'] = df['text'].apply(textstat.flesch_reading_ease)
df['word_count'] = df['text'].apply(textstat.lexicon_count)
df['lexical_diversity'] = df['cleaned_tokens'].apply(get_lexical_diversity)
```

```python
summary_table = df.groupby('platform')[
    ['sentiment_score', 'readability', 'lexical_diversity', 'word_count']
].mean().round(3)
summary_table
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='platform', y='sentiment_score', palette="Set2")
plt.title('Розподіл емоційності текстів (Sentiment Score)', fontsize=14)
plt.ylabel('Тональність (-1 до 1)')
plt.xlabel('Платформа')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.violinplot(data=df, x='platform', y='readability', palette="Set3")
plt.title('Читабельність текстів (Flesch Reading Ease)', fontsize=14)
plt.ylabel('Індекс читабельності (Вище = простіше)')
plt.xlabel('Платформа')
plt.tight_layout()
plt.show()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

tfidf = TfidfVectorizer(max_df=0.85, min_df=2)
grouped_text = df.groupby('platform')['cleaned_tokens'].apply(lambda x: ' '.join(x.astype(str)))

tfidf_matrix = tfidf.fit_transform(grouped_text)
feature_names = np.array(tfidf.get_feature_names_out())

for i, platform in enumerate(grouped_text.index):
    platform_vector = tfidf_matrix[i].toarray()[0]

    top_indices = platform_vector.argsort()[-10:][::-1]
    top_words = feature_names[top_indices]

    print(f"\nПлатформа: {platform}")
    print(", ".join(top_words))
```
